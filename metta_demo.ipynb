```python
from hyperon import MeTTa

def main():
    metta = MeTTa()
    print("--- Loading Knowledge Atoms into Atomspace ---")
    metta.run('''
        (Parent Abraham Isaac)
        (Parent Isaac Jacob)
        (Parent Jacob Joseph)
    ''')

    print("--- Loading Logical Reasoning Rules ---")
    metta.run('''
        (= (Grandparent $x $y)
           (match &self (and (Parent $x $z) (Parent $z $y)) $x))

        (= (Ancestor $x $y)
           (match &self (Parent $x $y) True))

        (= (Ancestor $x $y)
           (match &self (Parent $x $z) (Ancestor $z $y)))
    ''')

    print("--- Executing Queries ---\n")

    print("Query 1: Who is the Grandparent of Jacob?")
    query_1 = metta.run('! (Grandparent $who Jacob)')
    print(f"Result: {query_1}\n")

    print("Query 2: Is Abraham an ancestor of Joseph?")
    query_2 = metta.run('! (Ancestor Abraham Joseph)')
    print(f"Result: {query_2}\n")

    print("Query 3: List all parents stored in the Atomspace:")
    query_3 = metta.run('! (match &self (Parent $parent $child) $parent)')
    print(f"Result: {query_3}")

main()
```

```python
import pandas as pd
from hyperon import MeTTa

def main():
    metta = MeTTa()

    print("--- 1. Establishing Strict Type Enforcements ---")

    metta.run('''
        (: Person Type)
        (: Parent (-> Person Person Atom))
        (: Generation (-> Person Number Atom))
    ''')

    print("--- 2. Seeding Enhanced Knowledge Graph Atoms ---")

    metta.run('''
        (Parent Abraham Isaac)
        (Parent Isaac Jacob)
        (Parent Jacob Joseph)
        (Parent Joseph Ephraim)

        (Generation Abraham 1)
        (Generation Isaac 2)
        (Generation Jacob 3)
        (Generation Joseph 4)
        (Generation Ephraim 5)
    ''')

    print("--- 3. Injecting Transitive Ancestry Logic ---")
    metta.run('''
        (= (Grandparent $x $y)
           (match &self (and (Parent $x $z) (Parent $z $y)) $x))

        (= (Ancestor $x $y)
           (match &self (Parent $x $y) True))

        (= (Ancestor $x $y)
           (match &self (Parent $x $z) (Ancestor $z $y)))
    ''')

    print("--- 4. Extracting Data and Mapping to Python Analytics ---")

    raw_relations = metta.run('! (match &self (Parent $p $c) ($p $c))')


    cleaned_data = []
    if raw_relations and isinstance(raw_relations, list):

        for atom_group in raw_relations[0]:
            parent_name = str(atom_group.get_children()[0])
            child_name = str(atom_group.get_children()[1])
            cleaned_data.append({"Parent": parent_name, "Child": child_name})


    df = pd.DataFrame(cleaned_data)

    print("\n[SUCCESS] Extracted DataFrame generated from MeTTa Engine:")
    print(df.to_string(index=False))

    test_ancestor = metta.run('! (Ancestor Abraham Ephraim)')
    print(f"\nDeep Lineage Validation (Is Abraham the Ancestor of Ephraim?): {test_ancestor}")

if __name__ == "__main__":
    main()
```

```python
from hyperon import MeTTa

def main():
    metta = MeTTa()

    print("--- 1. Loading Expanded Raw Data Atoms ---")
    # Using only simple fact assertions just like before
    metta.run('''
        (Parent Abraham Isaac)
        (Parent Isaac Jacob)
        (Parent Jacob Joseph)
        (Parent Jacob Benjamin)
    ''')

    print("--- 2. Creating New Rules with Basic Matching ---")
    # Using only standard '=' and 'match' statements to find siblings
    metta.run('''
        (= (Grandparent $x $y)
           (match &self (and (Parent $x $z) (Parent $z $y)) $x))

        (= (Sibling $x $y)
           (match &self (and (Parent $parent $x) (Parent $parent $y)) $y))
    ''')

    print("--- 3. Executing New Match Patterns ---\n")

    # Query 1: Find the sibling of Joseph
    print("Query 1: Who is the sibling of Joseph?")
    query_1 = metta.run('! (Sibling Joseph $who)')
    print(f"Result: {query_1}\n")
    # Expected output shows Benjamin (and Joseph himself, due to pure pattern substitution matching)

    # Query 2: Find all children of Jacob using standard matching
    print("Query 2: List all direct children of Jacob:")
    query_2 = metta.run('! (match &self (Parent Jacob $child) $child)')
    print(f"Result: {query_2}")

if __name__ == "__main__":
    main()
```

```python
from hyperon import MeTTa

def main():
    metta = MeTTa()

    print("---  Loading E-Commerce Data Atoms ---")
    metta.run('''
        (HasCategory ItemA Electronics)
        (HasCategory ItemB Electronics)
        (HasCategory ItemC Books)
        (HasCategory ItemD Electronics)
    ''')

    print("--- Injecting Recommendation Logic (Shared Categories) ---")
    metta.run('''
        (= (RecommendSimilar $x $y)
           (match &self (and (HasCategory $x $category) (HasCategory $y $category)) $y))
    ''')

    print("Query: What items are recommended if a user views ItemA?")
    result = metta.run('! (RecommendSimilar ItemA $item)')
    print(f"Result: {result}\n")

if __name__ == "__main__":
    main()
```

```python
from hyperon import MeTTa

def main():
    metta = MeTTa()

    print("---  Loading Robotics State Atoms ---")
    metta.run('''
        (TriggerAction LowBattery ReturnToDock)
        (TriggerAction ObstacleDetected Brake)
        (TriggerAction PathClear MoveForward)
    ''')

    print("--- Injecting Simple Routing Rule ---")
    metta.run('''
        (= (GetAgentAction $current_state)
           (match &self (TriggerAction $current_state $action) $action))
    ''')

    print("Query: What action should the robot take if an Obstacle is Detected?")
    result = metta.run('! (GetAgentAction ObstacleDetected)')
    print(f"Result: {result}")

if __name__ == "__main__":
    main()
```

```python
from hyperon import MeTTa

def main():
    metta = MeTTa()

    print("---  Loading RBAC Security Atoms ---")
    metta.run('''
        (UserRole Alice Admin)
        (UserRole Bob Employee)
        (RolePermission Admin ReadPermission)
        (RolePermission Admin WritePermission)
        (RolePermission Employee ReadPermission)
    ''')

    print("--- Injecting Access Control Rules ---")
    metta.run('''
        (= (HasAccess $user $perm)
           (match &self (and (UserRole $user $role) (RolePermission $role $perm)) True))
    ''')

    print("Query: Does Bob have WritePermission?")
    result1 = metta.run('! (HasAccess Bob WritePermission)')
    print(f"Result (Bob): {result1}")

    print("Query: Does Alice have WritePermission?")
    result2 = metta.run('! (HasAccess Alice WritePermission)')
    print(f"Result (Alice): {result2}\n")

if __name__ == "__main__":
    main()
```
